```python
import torch
import torch.nn as nn
import torchvision
from torchvision import transforms, datasets
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm.notebook import tqdm
```

```python
class FlattenTransform:
    def __init__(self, start_dim=1):
        self.start_dim = start_dim

    def __call__(self, x):
        return x.flatten(start_dim=self.start_dim)

transform = transforms.Compose([
    transforms.ToTensor(),  # Converts image to PyTorch tensor
    transforms.Normalize((0.5,), (0.5,)),  # Normalizes the image (mean, std)
    # FlattenTranform()
])
```

```python
# Download and load the training dataset
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# Download and load the test dataset
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
```

```python

# Create DataLoader for training data
train_loader = DataLoader(dataset=train_dataset, batch_size=32, shuffle=True)

# Create DataLoader for test data
test_loader = DataLoader(dataset=test_dataset, batch_size=32, shuffle=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Model(nn.Module):
  def __init__(self):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(784, 128),
        nn.ReLU(),

        nn.Linear(128, 128),
        nn.ReLU(),

        nn.Linear(128, 10),
        nn.Softmax(dim=1)
    )

  def forward(self, x):
    return self.network(x)

model = Model()
model.to(device)
```

```python
critirion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.003)
```

```python
epochs = 20

def one_hot(y):
  one_hot_y = torch.zeros((y.size(0), 10))
  one_hot_y[torch.arange(y.size(0)), y] = 1
  return one_hot_y

for epoch in range(epochs):
  correct = 0
  total = 0
  for images, labels in tqdm(train_loader, colour="white"):
    images, labels = images.to(device), labels.to(device)
    images = torch.flatten(images, start_dim=1)
    # print(images.shape)
    # break
    y_pred = model(images)
    y = one_hot(labels)
    loss = critirion(y_pred, y)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    correct += (torch.max(y_pred, 1)[1] == torch.max(y, 1)[1]).sum().item()
    total += y.size(0)
    # print(y.size(0))
    # .item()
  print(f"epoch {epoch+1}/{epochs} accuracy: {100 * correct / total:.4f} %")

```

```python
def softmax1d(z):
  z = z - np.max(z, keepdims=True)
  exp_z = np.exp(z)
  return exp_z / np.sum(exp_z, keepdims=True)

def softmax(z, val=False):

  return np.apply_along_axis(softmax1d, axis=1, arr=z)
```

```python
class NeuralNetwork_wrelu:
  def __init__(self, input_shape, *layers):
    self.layers = layers
    units_in = input_shape
    self.history = {layer: {param: {"shape": None, "average": []} for param in ["w", "b", "z", "a", "db", "dw", "dz"]} for layer in range(len(self.layers))}
    for layer in self.layers:
      units_out = layer.units
      self.init_layer_params(layer, units_in, units_out)
      units_in = layer.units



  def init_layer_params(self, layer, input, output, relu=True):
    layer.w = np.random.randn(input, output) * 0.5
    layer.b = np.ones((1, output)) * 0.1 if relu else np.random.randn(1, output)
    layer.dz = np.zeros_like(layer.b)
    layer.dw = np.zeros_like(layer.w)
    layer.db = np.zeros_like(layer.b)

  def forward_prop(self, x, val=False):
    a = x
    for i, layer in enumerate(self.layers):

      layer.output = a @ layer.w + layer.b
      layer.activated = layer.activation.formula(layer.output, val=val)

      if layer.dropout is not None:
        layer.activated *= layer.dropout
      a = layer.activated
      self.history[i]["w"]["average"].append(np.average(layer.w))
      self.history[i]["b"]["average"].append(np.average(layer.b))
      self.history[i]["z"]["average"].append(np.average(layer.output))
      self.history[i]["a"]["average"].append(np.average(layer.activated))
      self.history[i]["dw"]["average"].append(np.average(layer.dw))
      self.history[i]["db"]["average"].append(np.average(layer.db))
      self.history[i]["dz"]["average"].append(np.average(layer.dz))

    return a

  def back_prop(self, x, y_hat, y):
    m = 1/x.shape[0]


    dz = (y_hat - y)
    for l in range(len(self.layers)-1, 0, -1):
      layer = self.layers[l]
      previous_layer = self.layers[l-1]
      a = previous_layer.output
      layer.dz = dz
      layer.dw = a.T @ dz * m

      layer.db = np.sum(dz, axis=0)
      dz = dz @ layer.w.T * previous_layer.activation.derivative(a)

    self.layers[0].dw = x.T @ dz * m
    self.layers[0].db = np.sum(dz, axis=0)




  def update_params(self, weight_decay=0.005):
    for layer in self.layers:
      layer.w -= self.lr * layer.dw
      layer.b -= self.lr * layer.db


  def gradient_descent(self, x, y, epochs=20, batch_size=32, learning_rate=0.01, val_split=0.1):
    self.lr = learning_rate

    s = x.shape[0]
    batchs = int(s * (1 - val_split) / batch_size)


    for epoch in range(1, epochs+1):


      threshold = np.random.rand(1)[0] * (1-val_split)
      val_start = round(s * threshold)
      val_end = min(s, round(s * (threshold + val_split)))
      x_val = x[val_start:val_end, :]
      y_val = y[val_start:val_end, :]

      if val_start:
        x_train = np.vstack((x[:val_start, :], x[val_end:, :]))
        y_train = np.vstack((y[:val_start, :], y[val_end:, :]))
      else:
        x_train = x[val_end:, :]
        y_train = y[val_end:, :]

      for batch in tqdm(range(batchs-1), colour="white"):
        # if not batch % 100 or batch > 940:
        #     print(batch)
        a = batch * batch_size
        b = min(s, (batch+1)*batch_size)
        x_batch = x_train[a:b, :]
        y_batch = y_train[a:b, :]

        y_hat = self.forward_prop(x_batch)
        self.back_prop(x_batch, y_hat, y_batch)
        self.update_params()
      print(f"[info] epoch {epoch}/{epochs}: val_accuracy: {round(self.evaluate(x_val, y_val), 4)} %")

    for i in range(len(self.layers)):
      self.history[i]["w"]["shape"] = self.layers[i].w.shape
      self.history[i]["b"]["shape"] = self.layers[i].b.shape
      self.history[i]["z"]["shape"] = self.layers[i].output.shape
      self.history[i]["a"]["shape"] = self.layers[i].activated.shape
      self.history[i]["dw"]["shape"] = self.layers[i].dw.shape
      self.history[i]["db"]["shape"] = self.layers[i].db.shape
      self.history[i]["dz"]["shape"] = self.layers[i].dz.shape


  def evaluate(self, x, y):

    prediction = self.forward_prop(x, val=True)
    return 100 * np.sum(np.argmax(prediction, axis=1) == np.argmax(y, axis=1)) / y.shape[0]


class Activation:
  def __init__(self, formula, derivative, value=None):
    self.value = value
    self.formula = formula
    self.derivative = derivative

class Layer:
  Relu = Activation(
        formula=lambda z, val=False: np.maximum(0, z),
        derivative=lambda z: z > 0
  )


  def __init__(self, units, activation=Relu, input=False, output=False, dropout_rate=None):

    self.dropout = np.random.choice([0, 1/(1-dropout_rate)], size=units, p=[dropout_rate, 1-dropout_rate]).reshape(1, units) if dropout_rate else None
    self.units = units
    self.input=input
    if not input and not output:
      self.activation = activation
    # self.ouput = np.zeros(1, units)
    if output:
      self.activation = activation
    self.w = None
    self.b = None
    self.ouput = None
    self.activated = None
```
